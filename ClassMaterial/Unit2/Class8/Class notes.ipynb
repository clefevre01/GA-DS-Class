```python
sales = pd.read_excel('/Users/cameronlefevre/Test-Repo/ClassMaterial/Unit2/data/superstore.xls', parse_dates=['Order Date'])
```

```python
sales.head()
```

```python
sales = sales.sort_values(by='Order Date')
```

```python
profits = sales.groupby('Order Date')[['Profit']].sum()
```

```python
profits['yesterday'] = profits.shift()
```

```python
profits
```

```python
profits['Profit'].shift(2)
```

```python
profits['profit2daysago'] = profits['Profit'].shift(2)
```

```python
profits.head()
```

```python
profits['Profit'].pct_change()
```

```python
profits['Profit'].rolling(7).mean().head(10)
```

```python
profits['7daystd'] = profits['Profit'].rolling(7).std()
```
