```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/cameronlefevre/Test-Repo/ClassMaterial/Unit2/data/restaurants.csv', parse_dates=['visit_date'])
```

```python
df.head()
```

```python
df['Weekend'] = np.where((df.day_of_week == 'Saturday') | (df.day_of_week == 'Sunday'), 'Weekend', 'Weekday')
```

```python
df.head()
```

```python
df['reserve_visitors'] > df['reserve_visitors'].quantile(.75)
```

```python
conditions = [
    df['reserve_visitors'] > df['reserve_visitors'].quantile(.75),
    (df['reserve_visitors'] <= df['reserve_visitors'].quantile(.75)) & (df['reserve_visitors'] > df['reserve_visitors'].quantile(.25)),
    df['reserve_visitors'] <= df['reserve_visitors'].quantile(.25),
]
results = ['High', 'Medium', 'Low']

df['ReserveQ'] = np.select(conditions, results, 'Other')

```

```python
df['time'] = (df['visit_date'] - df['visit_date'].min()).dt.days
```
